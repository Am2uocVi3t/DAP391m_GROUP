```python
# --- (Dán danh sách PROVINCES_COORDINATES từ câu trả lời trước vào đây) ---
PROVINCES_COORDINATES = {
    "An Giang": {"latitude": 10.5240, "longitude": 105.1248},
    "Ba Ria - Vung Tau": {"latitude": 10.5389, "longitude": 107.1986},
    "Bac Giang": {"latitude": 21.3038, "longitude": 106.1831},
    "Bac Kan": {"latitude": 22.2488, "longitude": 105.8237},
    "Bac Lieu": {"latitude": 9.2942, "longitude": 105.5048},
    "Bac Ninh": {"latitude": 21.1381, "longitude": 106.0751},
    "Ben Tre": {"latitude": 10.1259, "longitude": 106.4405},
    "Binh Dinh": {"latitude": 14.1209, "longitude": 108.9903},
    "Binh Duong": {"latitude": 11.1582, "longitude": 106.6386},
    "Binh Phuoc": {"latitude": 11.7239, "longitude": 106.8409},
    "Binh Thuan": {"latitude": 11.1187, "longitude": 108.2279},
    "Ca Mau": {"latitude": 9.1730, "longitude": 105.1500},
    "Can Tho": {"latitude": 10.0452, "longitude": 105.7469},
    "Cao Bang": {"latitude": 22.6691, "longitude": 106.2601},
    "Da Nang": {"latitude": 16.0479, "longitude": 108.2208},
    "Dak Lak": {"latitude": 12.6693, "longitude": 108.0436},
    "Dak Nong": {"latitude": 12.2276, "longitude": 107.6955},
    "Dien Bien": {"latitude": 21.3840, "longitude": 103.0168},
    "Dong Nai": {"latitude": 10.9519, "longitude": 106.8242},
    "Dong Thap": {"latitude": 10.4598, "longitude": 105.6367},
    "Gia Lai": {"latitude": 13.9780, "longitude": 108.0030},
    "Ha Giang": {"latitude": 22.8103, "longitude": 104.9804},
    "Ha Nam": {"latitude": 20.5467, "longitude": 105.9073},
    "Ha Noi": {"latitude": 21.0285, "longitude": 105.8542},
    "Ha Tinh": {"latitude": 18.3425, "longitude": 105.9031},
    "Hai Duong": {"latitude": 20.9372, "longitude": 106.3043},
    "Hai Phong": {"latitude": 20.8449, "longitude": 106.6881},
    "Hau Giang": {"latitude": 9.7788, "longitude": 105.4680},
    "Hoa Binh": {"latitude": 20.8192, "longitude": 105.3371},
    "Hung Yen": {"latitude": 20.6477, "longitude": 106.0543},
    "Khanh Hoa": {"latitude": 12.2388, "longitude": 109.1967},
    "Kien Giang": {"latitude": 9.9995, "longitude": 105.1050},
    "Kon Tum": {"latitude": 14.3501, "longitude": 108.0015},
    "Lai Chau": {"latitude": 22.3985, "longitude": 103.4591},
    "Lam Dong": {"latitude": 11.9404, "longitude": 108.4583},
    "Lang Son": {"latitude": 21.8503, "longitude": 106.7592},
    "Lao Cai": {"latitude": 22.4861, "longitude": 103.9700},
    "Long An": {"latitude": 10.6719, "longitude": 106.3361},
    "Nam Dinh": {"latitude": 20.4276, "longitude": 106.1704},
    "Nghe An": {"latitude": 18.6756, "longitude": 105.6859},
    "Ninh Binh": {"latitude": 20.2512, "longitude": 105.9763},
    "Ninh Thuan": {"latitude": 11.5733, "longitude": 108.9878},
    "Phu Tho": {"latitude": 21.3274, "longitude": 105.4034},
    "Phu Yen": {"latitude": 13.0972, "longitude": 109.2988},
    "Quang Binh": {"latitude": 17.4723, "longitude": 106.6103},
    "Quang Nam": {"latitude": 15.5726, "longitude": 108.4721},
    "Quang Ngai": {"latitude": 15.1202, "longitude": 108.7920},
    "Quang Ninh": {"latitude": 20.9544, "longitude": 107.0740},
    "Quang Tri": {"latitude": 16.8048, "longitude": 107.1859},
    "Soc Trang": {"latitude": 9.6037, "longitude": 105.9721},
    "Son La": {"latitude": 21.3257, "longitude": 103.9126},
    "Tay Ninh": {"latitude": 11.3094, "longitude": 106.0989},
    "Thai Binh": {"latitude": 20.4487, "longitude": 106.3390},
    "Thai Nguyen": {"latitude": 21.5906, "longitude": 105.8440},
    "Thanh Hoa": {"latitude": 19.8032, "longitude": 105.7760},
    "Thua Thien Hue": {"latitude": 16.4637, "longitude": 107.5909},
    "Tien Giang": {"latitude": 10.3566, "longitude": 106.3481},
    "Ho Chi Minh City": {"latitude": 10.7769, "longitude": 106.7009},
    "Tra Vinh": {"latitude": 9.9497, "longitude": 106.3412},
    "Tuyen Quang": {"latitude": 21.8067, "longitude": 105.2186},
    "Vinh Long": {"latitude": 10.2516, "longitude": 105.9712},
    "Vinh Phuc": {"latitude": 21.3097, "longitude": 105.6040},
    "Yen Bai": {"latitude": 21.7242, "longitude": 104.9055},
}
# --- END PROVINCES_COORDINATES ---

# DANH SÁCH TẤT CẢ CÁC BIẾN HOURLY TỪ OPEN-METEO (CẬP NHẬT THÁNG 6/2024 - KIỂM TRA LẠI TRÊN DOCS)
ALL_HOURLY_PARAMS_FULL_LIST = [
    "temperature_2m", "relative_humidity_2m", "dew_point_2m", "apparent_temperature",
    "precipitation_probability", "precipitation", "rain", "showers", "snowfall", "snow_depth",
    "weather_code", "pressure_msl", "surface_pressure", "cloud_cover", "cloud_cover_low",
    "cloud_cover_mid", "cloud_cover_high", "visibility", "evapotranspiration",
    "et0_fao_evapotranspiration", "vapour_pressure_deficit", "wind_speed_10m", "wind_speed_80m",
    "wind_speed_100m", "wind_speed_120m", "wind_speed_180m", "wind_direction_10m",
    "wind_direction_80m", "wind_direction_100m", "wind_direction_120m", "wind_direction_180m",
    "wind_gusts_10m", "temperature_80m", "temperature_120m", "temperature_180m",
    "soil_temperature_0cm", "soil_temperature_6cm", "soil_temperature_18cm", "soil_temperature_54cm",
    "soil_moisture_0_to_1cm", "soil_moisture_1_to_3cm", "soil_moisture_3_to_9cm",
    "soil_moisture_9_to_27cm", "soil_moisture_27_to_81cm", "shortwave_radiation",
    "direct_radiation", "diffuse_radiation", "direct_normal_irradiance",
    "terrestrial_radiation", "shortwave_radiation_instant", "direct_radiation_instant",
    "diffuse_radiation_instant", "direct_normal_irradiance_instant",
    "terrestrial_radiation_instant", "uv_index", "uv_index_clear_sky", "cape",
    "lifted_index", "is_day", "sunshine_duration"
    # Các biến như global_tilted_irradiance, aqi, pollen, river, wave đã được loại bỏ
    # vì chúng cần cấu hình API bổ sung hoặc model cụ thể.
]

# Sử dụng danh sách đã được đơn giản hóa này
REQUESTED_HOURLY_PARAMS = ALL_HOURLY_PARAMS_FULL_LIST.copy() # Tạo bản copy để không ảnh hưởng list gốc

def get_weather_data_for_location(latitude, longitude, start_date, end_date, location_name, hourly_params_list):
    """
    Lấy dữ liệu thời tiết hàng giờ cho một địa điểm cụ thể với danh sách biến được chỉ định.
    """
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "hourly": ",".join(hourly_params_list), # Sử dụng danh sách biến được truyền vào
        "start_date": start_date.strftime("%Y-%m-%d"),
        "end_date": end_date.strftime("%Y-%m-%d"),
        "timezone": "Asia/Ho_Chi_Minh"
    }
    try:
        response = requests.get(url, params=params, timeout=30) # Thêm timeout cho request
        response.raise_for_status()
        data = response.json()

        if 'hourly' not in data or not data['hourly']:
            print(f"Không có dữ liệu 'hourly' hoặc dữ liệu rỗng cho {location_name} với các params đã chọn. Response: {data}")
            return None

        hourly_data = data['hourly']
        lengths = {key: len(value) for key, value in hourly_data.items()}
        if len(set(lengths.values())) > 1:
            # Nếu độ dài không nhất quán, chỉ lấy các key có độ dài phổ biến nhất (thường là 'time')
            if 'time' in lengths:
                target_length = lengths['time']
                consistent_hourly_data = {}
                for key, value in hourly_data.items():
                    if len(value) == target_length:
                        consistent_hourly_data[key] = value
                    else:
                        print(f"  Cảnh báo: Bỏ qua biến '{key}' cho {location_name} do độ dài không nhất quán ({len(value)} vs {target_length}).")
                hourly_data = consistent_hourly_data
                if not hourly_data or 'time' not in hourly_data: # Nếu sau khi lọc không còn time thì lỗi
                     print(f"Dữ liệu không nhất quán và không có 'time' sau khi lọc cho {location_name}.")
                     return None
            else: # Không có 'time' để làm tham chiếu
                print(f"Các trường dữ liệu hourly không cùng độ dài và không có 'time' làm tham chiếu cho {location_name}. Lengths: {lengths}")
                return None


        df = pd.DataFrame(hourly_data)
        df['time'] = pd.to_datetime(df['time'])
        df['province'] = location_name
        df['latitude_query'] = latitude
        df['longitude_query'] = longitude
        return df

    except requests.exceptions.Timeout:
        print(f"Lỗi Timeout khi gọi API cho {location_name}.")
        return None
    except requests.exceptions.RequestException as e:
        print(f"Lỗi khi gọi API cho {location_name}: {e}")
        return None
    except KeyError as e:
        print(f"Lỗi KeyError khi xử lý dữ liệu cho {location_name} (thiếu key {e}). Response: {data}")
        return None
    except Exception as e:
        print(f"Lỗi không xác định khi xử lý {location_name}: {e}")
        return None

```

```python
def call_API(city, days):
    end_date = datetime.now()
    start_date = end_date - days

    print(f"Thu thập dữ liệu từ {start_date.strftime('%Y-%m-%d')} đến {end_date.strftime('%Y-%m-%d')} cho 63 tỉnh thành.")
    print(f"Yêu cầu {len(REQUESTED_HOURLY_PARAMS)} biến hàng giờ.")

    all_weather_data = []
    total_provinces = len(PROVINCES_COORDINATES)
    processed_count = 0
    request_delay = 0.2 # Giây, tăng nếu gặp lỗi 429

    for province_name, coords in PROVINCES_COORDINATES.items():
        processed_count += 1
        print(f"Đang lấy dữ liệu cho: {province_name} ({coords['latitude']}, {coords['longitude']}) - [{processed_count}/{total_provinces}]")
        df_province = get_weather_data_for_location(
            coords['latitude'],
            coords['longitude'],
            start_date,
            end_date,
            province_name,
            REQUESTED_HOURLY_PARAMS # Truyền danh sách biến vào hàm
        )
        if df_province is not None:
            all_weather_data.append(df_province)

        if processed_count < total_provinces: # Không delay sau request cuối cùng
            time.sleep(request_delay)
        print("-" * 40)

    if not all_weather_data:
        print("Không thu thập được dữ liệu nào. Vui lòng kiểm tra lại.")
    else:
        final_df = pd.concat(all_weather_data, ignore_index=True)

        cols_to_front = ['province', 'latitude_query', 'longitude_query', 'time']
        other_cols = [col for col in final_df.columns if col not in cols_to_front]
        final_df = final_df[cols_to_front + other_cols]

        output_filename = f"vietnam_63provinces_ALL_VARS_weather_sample_{start_date.strftime('%Y%m%d')}_{end_date.strftime('%Y%m%d')}.csv"
        final_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
        print(f"\nĐã lưu dữ liệu vào file: {output_filename}")
        print(f"Tổng số bản ghi: {len(final_df)}")
        print(f"Tổng số cột (biến): {len(final_df.columns)}")
        print(f"Số tỉnh/thành phố có dữ liệu: {final_df['province'].nunique()}")
        print("Xem qua vài dòng đầu của dữ liệu (chỉ một vài cột đầu):")
        print(final_df.iloc[:, :7].head()) # In 7 cột đầu
```
