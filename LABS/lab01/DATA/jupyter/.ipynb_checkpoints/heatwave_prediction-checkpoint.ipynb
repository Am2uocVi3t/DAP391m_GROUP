```python
# --- (Dán danh sách PROVINCES_COORDINATES từ câu trả lời trước vào đây) ---
PROVINCES_COORDINATES = {
    "An Giang": {"latitude": 10.5240, "longitude": 105.1248},
    "Ba Ria - Vung Tau": {"latitude": 10.5389, "longitude": 107.1986},
    "Bac Giang": {"latitude": 21.3038, "longitude": 106.1831},
    "Bac Kan": {"latitude": 22.2488, "longitude": 105.8237},
    "Bac Lieu": {"latitude": 9.2942, "longitude": 105.5048},
    "Bac Ninh": {"latitude": 21.1381, "longitude": 106.0751},
    "Ben Tre": {"latitude": 10.1259, "longitude": 106.4405},
    "Binh Dinh": {"latitude": 14.1209, "longitude": 108.9903},
    "Binh Duong": {"latitude": 11.1582, "longitude": 106.6386},
    "Binh Phuoc": {"latitude": 11.7239, "longitude": 106.8409},
    "Binh Thuan": {"latitude": 11.1187, "longitude": 108.2279},
    "Ca Mau": {"latitude": 9.1730, "longitude": 105.1500},
    "Can Tho": {"latitude": 10.0452, "longitude": 105.7469},
    "Cao Bang": {"latitude": 22.6691, "longitude": 106.2601},
    "Da Nang": {"latitude": 16.0479, "longitude": 108.2208},
    "Dak Lak": {"latitude": 12.6693, "longitude": 108.0436},
    "Dak Nong": {"latitude": 12.2276, "longitude": 107.6955},
    "Dien Bien": {"latitude": 21.3840, "longitude": 103.0168},
    "Dong Nai": {"latitude": 10.9519, "longitude": 106.8242},
    "Dong Thap": {"latitude": 10.4598, "longitude": 105.6367},
    "Gia Lai": {"latitude": 13.9780, "longitude": 108.0030},
    "Ha Giang": {"latitude": 22.8103, "longitude": 104.9804},
    "Ha Nam": {"latitude": 20.5467, "longitude": 105.9073},
    "Ha Noi": {"latitude": 21.0285, "longitude": 105.8542},
    "Ha Tinh": {"latitude": 18.3425, "longitude": 105.9031},
    "Hai Duong": {"latitude": 20.9372, "longitude": 106.3043},
    "Hai Phong": {"latitude": 20.8449, "longitude": 106.6881},
    "Hau Giang": {"latitude": 9.7788, "longitude": 105.4680},
    "Hoa Binh": {"latitude": 20.8192, "longitude": 105.3371},
    "Hung Yen": {"latitude": 20.6477, "longitude": 106.0543},
    "Khanh Hoa": {"latitude": 12.2388, "longitude": 109.1967},
    "Kien Giang": {"latitude": 9.9995, "longitude": 105.1050},
    "Kon Tum": {"latitude": 14.3501, "longitude": 108.0015},
    "Lai Chau": {"latitude": 22.3985, "longitude": 103.4591},
    "Lam Dong": {"latitude": 11.9404, "longitude": 108.4583},
    "Lang Son": {"latitude": 21.8503, "longitude": 106.7592},
    "Lao Cai": {"latitude": 22.4861, "longitude": 103.9700},
    "Long An": {"latitude": 10.6719, "longitude": 106.3361},
    "Nam Dinh": {"latitude": 20.4276, "longitude": 106.1704},
    "Nghe An": {"latitude": 18.6756, "longitude": 105.6859},
    "Ninh Binh": {"latitude": 20.2512, "longitude": 105.9763},
    "Ninh Thuan": {"latitude": 11.5733, "longitude": 108.9878},
    "Phu Tho": {"latitude": 21.3274, "longitude": 105.4034},
    "Phu Yen": {"latitude": 13.0972, "longitude": 109.2988},
    "Quang Binh": {"latitude": 17.4723, "longitude": 106.6103},
    "Quang Nam": {"latitude": 15.5726, "longitude": 108.4721},
    "Quang Ngai": {"latitude": 15.1202, "longitude": 108.7920},
    "Quang Ninh": {"latitude": 20.9544, "longitude": 107.0740},
    "Quang Tri": {"latitude": 16.8048, "longitude": 107.1859},
    "Soc Trang": {"latitude": 9.6037, "longitude": 105.9721},
    "Son La": {"latitude": 21.3257, "longitude": 103.9126},
    "Tay Ninh": {"latitude": 11.3094, "longitude": 106.0989},
    "Thai Binh": {"latitude": 20.4487, "longitude": 106.3390},
    "Thai Nguyen": {"latitude": 21.5906, "longitude": 105.8440},
    "Thanh Hoa": {"latitude": 19.8032, "longitude": 105.7760},
    "Thua Thien Hue": {"latitude": 16.4637, "longitude": 107.5909},
    "Tien Giang": {"latitude": 10.3566, "longitude": 106.3481},
    "Ho Chi Minh City": {"latitude": 10.7769, "longitude": 106.7009},
    "Tra Vinh": {"latitude": 9.9497, "longitude": 106.3412},
    "Tuyen Quang": {"latitude": 21.8067, "longitude": 105.2186},
    "Vinh Long": {"latitude": 10.2516, "longitude": 105.9712},
    "Vinh Phuc": {"latitude": 21.3097, "longitude": 105.6040},
    "Yen Bai": {"latitude": 21.7242, "longitude": 104.9055},
}
# --- END PROVINCES_COORDINATES ---

# DANH SÁCH TẤT CẢ CÁC BIẾN HOURLY TỪ OPEN-METEO (CẬP NHẬT THÁNG 6/2024 - KIỂM TRA LẠI TRÊN DOCS)
ALL_HOURLY_PARAMS_FULL_LIST = [
    "temperature_2m", "relative_humidity_2m", "dew_point_2m", "apparent_temperature",
    "precipitation_probability", "precipitation", "rain", "showers", "snowfall", "snow_depth",
    "weather_code", "pressure_msl", "surface_pressure", "cloud_cover", "cloud_cover_low",
    "cloud_cover_mid", "cloud_cover_high", "visibility", "evapotranspiration",
    "et0_fao_evapotranspiration", "vapour_pressure_deficit", "wind_speed_10m", "wind_speed_80m",
    "wind_speed_100m", "wind_speed_120m", "wind_speed_180m", "wind_direction_10m",
    "wind_direction_80m", "wind_direction_100m", "wind_direction_120m", "wind_direction_180m",
    "wind_gusts_10m", "temperature_80m", "temperature_120m", "temperature_180m",
    "soil_temperature_0cm", "soil_temperature_6cm", "soil_temperature_18cm", "soil_temperature_54cm",
    "soil_moisture_0_to_1cm", "soil_moisture_1_to_3cm", "soil_moisture_3_to_9cm",
    "soil_moisture_9_to_27cm", "soil_moisture_27_to_81cm", "shortwave_radiation",
    "direct_radiation", "diffuse_radiation", "direct_normal_irradiance",
    "terrestrial_radiation", "shortwave_radiation_instant", "direct_radiation_instant",
    "diffuse_radiation_instant", "direct_normal_irradiance_instant",
    "terrestrial_radiation_instant", "uv_index", "uv_index_clear_sky", "cape",
    "lifted_index", "is_day", "sunshine_duration"
    # Các biến như global_tilted_irradiance, aqi, pollen, river, wave đã được loại bỏ
    # vì chúng cần cấu hình API bổ sung hoặc model cụ thể.
]

# Sử dụng danh sách đã được đơn giản hóa này
REQUESTED_HOURLY_PARAMS = ALL_HOURLY_PARAMS_FULL_LIST.copy() # Tạo bản copy để không ảnh hưởng list gốc

def get_weather_data_for_location(latitude, longitude, start_date, end_date, location_name, hourly_params_list):
    """
    Lấy dữ liệu thời tiết hàng giờ cho một địa điểm cụ thể với danh sách biến được chỉ định.
    """
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "hourly": ",".join(hourly_params_list), # Sử dụng danh sách biến được truyền vào
        "start_date": start_date.strftime("%Y-%m-%d"),
        "end_date": end_date.strftime("%Y-%m-%d"),
        "timezone": "Asia/Ho_Chi_Minh"
    }
    try:
        response = requests.get(url, params=params, timeout=30) # Thêm timeout cho request
        response.raise_for_status()
        data = response.json()

        if 'hourly' not in data or not data['hourly']:
            print(f"Không có dữ liệu 'hourly' hoặc dữ liệu rỗng cho {location_name} với các params đã chọn. Response: {data}")
            return None

        hourly_data = data['hourly']
        lengths = {key: len(value) for key, value in hourly_data.items()}
        if len(set(lengths.values())) > 1:
            # Nếu độ dài không nhất quán, chỉ lấy các key có độ dài phổ biến nhất (thường là 'time')
            if 'time' in lengths:
                target_length = lengths['time']
                consistent_hourly_data = {}
                for key, value in hourly_data.items():
                    if len(value) == target_length:
                        consistent_hourly_data[key] = value
                    else:
                        print(f"  Cảnh báo: Bỏ qua biến '{key}' cho {location_name} do độ dài không nhất quán ({len(value)} vs {target_length}).")
                hourly_data = consistent_hourly_data
                if not hourly_data or 'time' not in hourly_data: # Nếu sau khi lọc không còn time thì lỗi
                     print(f"Dữ liệu không nhất quán và không có 'time' sau khi lọc cho {location_name}.")
                     return None
            else: # Không có 'time' để làm tham chiếu
                print(f"Các trường dữ liệu hourly không cùng độ dài và không có 'time' làm tham chiếu cho {location_name}. Lengths: {lengths}")
                return None


        df = pd.DataFrame(hourly_data)
        df['time'] = pd.to_datetime(df['time'])
        df['province'] = location_name
        df['latitude_query'] = latitude
        df['longitude_query'] = longitude
        return df

    except requests.exceptions.Timeout:
        print(f"Lỗi Timeout khi gọi API cho {location_name}.")
        return None
    except requests.exceptions.RequestException as e:
        print(f"Lỗi khi gọi API cho {location_name}: {e}")
        return None
    except KeyError as e:
        print(f"Lỗi KeyError khi xử lý dữ liệu cho {location_name} (thiếu key {e}). Response: {data}")
        return None
    except Exception as e:
        print(f"Lỗi không xác định khi xử lý {location_name}: {e}")
        return None

if __name__ == "__main__":
    end_date = datetime.now()
    start_date = end_date - timedelta(days=7)

    print(f"Thu thập dữ liệu từ {start_date.strftime('%Y-%m-%d')} đến {end_date.strftime('%Y-%m-%d')} cho 63 tỉnh thành.")
    print(f"Yêu cầu {len(REQUESTED_HOURLY_PARAMS)} biến hàng giờ.")

    all_weather_data = []
    total_provinces = len(PROVINCES_COORDINATES)
    processed_count = 0
    request_delay = 0.2 # Giây, tăng nếu gặp lỗi 429

    for province_name, coords in PROVINCES_COORDINATES.items():
        processed_count += 1
        print(f"Đang lấy dữ liệu cho: {province_name} ({coords['latitude']}, {coords['longitude']}) - [{processed_count}/{total_provinces}]")
        df_province = get_weather_data_for_location(
            coords['latitude'],
            coords['longitude'],
            start_date,
            end_date,
            province_name,
            REQUESTED_HOURLY_PARAMS # Truyền danh sách biến vào hàm
        )
        if df_province is not None:
            all_weather_data.append(df_province)

        if processed_count < total_provinces: # Không delay sau request cuối cùng
            time.sleep(request_delay)
        print("-" * 40)

    if not all_weather_data:
        print("Không thu thập được dữ liệu nào. Vui lòng kiểm tra lại.")
    else:
        final_df = pd.concat(all_weather_data, ignore_index=True)

        cols_to_front = ['province', 'latitude_query', 'longitude_query', 'time']
        other_cols = [col for col in final_df.columns if col not in cols_to_front]
        final_df = final_df[cols_to_front + other_cols]

        output_filename = f"vietnam_63provinces_ALL_VARS_weather_sample_{start_date.strftime('%Y%m%d')}_{end_date.strftime('%Y%m%d')}.csv"
        final_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
        print(f"\nĐã lưu dữ liệu vào file: {output_filename}")
        print(f"Tổng số bản ghi: {len(final_df)}")
        print(f"Tổng số cột (biến): {len(final_df.columns)}")
        print(f"Số tỉnh/thành phố có dữ liệu: {final_df['province'].nunique()}")
        print("Xem qua vài dòng đầu của dữ liệu (chỉ một vài cột đầu):")
        print(final_df.iloc[:, :7].head()) # In 7 cột đầu
```

```python
# --- Định nghĩa hàm độc lập get_weather_data_for_province_id_standalone (Đã cập nhật) ---
def get_weather_data_for_province_id_standalone(province_id, num_days):
    # Sử dụng dictionary province_to_number được tạo ở ngoài hàm
    province_name = None
    # Tìm tên tỉnh từ province_id
    for name, id in province_to_number.items():
        if id == province_id:
            province_name = name
            break

    if province_name is None:
        print(f"Lỗi: Không tìm thấy tỉnh với ID = {province_id}.")
        print(f"Các ID tỉnh hợp lệ là: {list(province_to_number.values())}")
        return None

    # Lấy tọa độ của tỉnh
    if province_name not in PROVINCES_COORDINATES:
        print(f"Lỗi: Không tìm thấy tọa độ cho tỉnh '{province_name}' ({province_id=}).")
        return None

    coords = PROVINCES_COORDINATES[province_name]
    latitude = coords['latitude']
    longitude = coords['longitude']

    # Tính toán ngày bắt đầu và ngày kết thúc
    end_date = datetime.now()
    start_date = end_date - timedelta(days=num_days)

    print(f"Đang lấy dữ liệu thời tiết cho tỉnh '{province_name}' (ID: {province_id}) trong {num_days} ngày...")

    # --- Logic gọi API trực tiếp ---
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "hourly": ",".join(REQUESTED_HOURLY_PARAMS), # Sử dụng danh sách biến đã định nghĩa ở đầu
        "start_date": start_date.strftime("%Y-%m-%d"),
        "end_date": end_date.strftime("%Y-%m-%d"),
        "timezone": "Asia/Ho_Chi_Minh"
    }
    try:
        response = requests.get(url, params=params, timeout=30)
        response.raise_for_status()
        data = response.json()

        if 'hourly' not in data or not data['hourly']:
            print(f"Không có dữ liệu 'hourly' hoặc dữ liệu rỗng cho {province_name} với các params đã chọn. Response: {data}")
            return None

        hourly_data = data['hourly']
        lengths = {key: len(value) for key, value in hourly_data.items()}
        if len(set(lengths.values())) > 1:
            if 'time' in lengths:
                target_length = lengths['time']
                consistent_hourly_data = {}
                for key, value in hourly_data.items():
                    if len(value) == target_length:
                        consistent_hourly_data[key] = value
                    else:
                        print(f"  Cảnh báo: Bỏ qua biến '{key}' cho {province_name} do độ dài không nhất quán ({len(value)} vs {target_length}).")
                hourly_data = consistent_hourly_data
                if not hourly_data or 'time' not in hourly_data:
                     print(f"Dữ liệu không nhất quán và không có 'time' sau khi lọc cho {province_name}.")
                     return None
            else:
                print(f"Các trường dữ liệu hourly không cùng độ dài và không có 'time' làm tham chiếu cho {province_name}. Lengths: {lengths}")
                return None

        df = pd.DataFrame(hourly_data)
        df['time'] = pd.to_datetime(df['time'])
        # Giữ lại hai cột này tạm thời nếu cần xử lý thêm, nhưng sẽ bỏ đi trước khi trả về
        df['province'] = province_name
        df['province_id'] = province_id
        df['latitude_query'] = latitude
        df['longitude_query'] = longitude


        # --- THAY ĐỔI: Loại bỏ cột 'province' và 'province_id' ---
        # Sử dụng .drop() với axis=1 để xóa cột
        cols_to_drop = ['province', 'province_id']
        # Chỉ xóa nếu các cột tồn tại trong DataFrame
        existing_cols_to_drop = [col for col in cols_to_drop if col in df.columns]
        if existing_cols_to_drop:
            df = df.drop(columns=existing_cols_to_drop)
            # print(f"Đã loại bỏ các cột {existing_cols_to_drop} khỏi DataFrame trả về.") # Tùy chọn: in thông báo

        # Sắp xếp lại cột: giữ lại các cột nhận dạng khác ở đầu
        cols_to_front = ['latitude_query', 'longitude_query', 'time']
        # Lọc ra các cột còn lại sau khi bỏ province, province_id
        other_cols = [col for col in df.columns if col not in cols_to_front]
        df = df[cols_to_front + other_cols]


        return df

    except requests.exceptions.Timeout:
        print(f"Lỗi Timeout khi gọi API cho {province_name}.")
        return None
    except requests.exceptions.RequestException as e:
        print(f"Lỗi khi gọi API cho {province_name}: {e}")
        return None
    except KeyError as e:
        print(f"Lỗi KeyError khi xử lý dữ liệu cho {province_name} (thiếu key {e}). Response: {data}")
        return None
    except Exception as e:
        print(f"Lỗi không xác định khi xử lý {province_name}: {e}")
        return None
    # --- Kết thúc logic gọi API ---
```

```python
# --- 1. Cleaning data (Xử lý dữ liệu) ---
print("\n--- 1. Cleaning data ---")

# Chuyển cột 'time' sang định dạng datetime
df['time'] = pd.to_datetime(df['time'])

# Sắp xếp dữ liệu theo thời gian (quan trọng cho dữ liệu chuỗi thời gian)
df = df.sort_values(by='time').reset_index(drop=True)

# Xử lý giá trị bị thiếu (Null/NaN)
print(f"\nSố lượng giá trị thiếu trước khi xử lý:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
```

```python
# Chiến lược xử lý giá trị thiếu:
# - Đối với các cột số, có thể điền bằng median hoặc ffill/bfill cho time series
# - Đối với các cột có tỷ lệ thiếu quá cao, có thể xem xét loại bỏ
# - Một số cột như 'precipitation', 'snowfall' có thể NaN nghĩa là 0

# Ví dụ: Điền 0 cho các cột liên quan đến mưa/tuyết nếu NaN
precipitation_cols = ['precipitation', 'rain', 'showers', 'snowfall', 'snow_depth']
for col in precipitation_cols:
    if col in df.columns:
        df[col] = df[col].fillna(0)

# Sử dụng ffill (forward fill) cho các cột số khác để duy trì tính chất chuỗi thời gian
# (chọn các cột số không bao gồm các cột đã xử lý ở trên)
numeric_cols_to_impute = df.select_dtypes(exclude='object').columns
numeric_cols_to_impute = [col for col in numeric_cols_to_impute if col not in precipitation_cols]

for col in numeric_cols_to_impute:
    df[col] = df[col].ffill() # Forward fill
    df[col] = df[col].bfill() # Backward fill (để xử lý NaN ở đầu nếu có)

print(f"\nSố lượng giá trị thiếu sau khi xử lý:\n{df.isnull().sum()[df.isnull().sum() > 0]}") # Tổng số NaN còn lại
```

```python
# Xử lý outliers (Giá trị ngoại lai)
# Sử dụng phương pháp IQR cho một vài cột ví dụ. Trong thực tế, cần xem xét cẩn thận từng cột.
# Chúng ta sẽ không loại bỏ hoàn toàn mà có thể giới hạn (cap) chúng.
# Tuy nhiên, với dữ liệu thời tiết, một số "outliers" có thể là giá trị thực (ví dụ: gió giật mạnh).
# Tạm thời, chúng ta sẽ không xử lý outlier mạnh tay ở bước này để tránh mất thông tin quan trọng.
# Một cách tiếp cận đơn giản là hiển thị boxplot để quan sát.

# Loại bỏ các cột có giá trị không đổi (ví dụ: latitude_query, longitude_query nếu chỉ có 1 địa điểm)
cols_to_drop = []
for col in df.columns:
    if df[col].nunique() == 1:
        cols_to_drop.append(col)
if cols_to_drop:
    print(f"\nLoại bỏ các cột có giá trị không đổi: {cols_to_drop}")
    df = df.drop(columns=cols_to_drop)

print(f"Kích thước dữ liệu sau khi cleaning cơ bản: {df.shape}")
```

```python
# --- 4. Feature engineering (Kỹ thuật đặc trưng) ---
print("\n--- 4. Feature engineering ---")

# a. Tạo các đặc trưng từ cột 'time'
df['hour'] = df['time'].dt.hour
df['dayofweek'] = df['time'].dt.dayofweek # Monday=0, Sunday=6
df['month'] = df['time'].dt.month
df['dayofyear'] = df['time'].dt.dayofyear

# b. Tạo các đặc trưng trễ (lagged features) cho các biến quan trọng
# Ví dụ: nhiệt độ, độ ẩm của 1, 2, 3 giờ trước
lag_features = ['temperature_2m', 'apparent_temperature', 'relative_humidity_2m', 'wind_speed_10m']
for col in lag_features:
    for lag in [1, 2, 3]: # Trễ 1, 2, 3 giờ
        df[f'{col}_lag{lag}'] = df[col].shift(lag)

# c. Tạo các đặc trưng lăn (rolling features)
# Ví dụ: trung bình nhiệt độ trong 3 giờ, 6 giờ qua
for col in ['temperature_2m', 'apparent_temperature']:
    for window in [3, 6]: # Cửa sổ 3 giờ, 6 giờ
        df[f'{col}_roll_mean{window}h'] = df[col].rolling(window=window, min_periods=1).mean()
        df[f'{col}_roll_std{window}h'] = df[col].rolling(window=window, min_periods=1).std()

# d. Đặc trưng tương tác (Interaction features) - ví dụ đơn giản
# Nhiệt độ chênh lệch so với điểm sương (có thể liên quan đến cảm giác oi bức)
if 'temperature_2m' in df.columns and 'dew_point_2m' in df.columns:
    df['temp_dewpoint_diff'] = df['temperature_2m'] - df['dew_point_2m']

# Sau khi tạo lag/rolling features, sẽ có NaN ở đầu, cần xử lý
df = df.bfill().ffill() # Hoặc sử dụng SimpleImputer sau này trong pipeline

print(f"Kích thước dữ liệu sau Feature Engineering: {df.shape}")
print(f"Một vài cột mới được tạo:\n {df[['hour', 'temperature_2m_lag1', 'apparent_temperature_roll_mean3h']].head()}")
```

```python
# --- 5. Preprocessing để đưa vào mô hình ---
print("\n--- 5. Preprocessing ---")

# Tách biến mục tiêu (y) và các biến độc lập (X)
X = df.drop(columns=['is_heatwave', 'time']) # Loại bỏ 'time' vì đã trích xuất đặc trưng
y = df['is_heatwave']

# Xác định các cột số và cột hạng mục
numerical_features = X.select_dtypes(include=np.number).columns.tolist()
categorical_features = X.select_dtypes(exclude=np.number).columns.tolist() # Nên là rỗng nếu tất cả đã là số

# weather_code, hour, dayofweek, month là các biến hạng mục mặc dù có thể đang là số
# Chúng ta sẽ coi chúng là hạng mục để OneHotEncode
potential_cat_from_num = ['weather_code', 'hour', 'dayofweek', 'month']
for p_cat in potential_cat_from_num:
    if p_cat in numerical_features:
        numerical_features.remove(p_cat)
        if p_cat not in categorical_features:
             categorical_features.append(p_cat)

print(f"Numerical features: {numerical_features[:5]} ... ({len(numerical_features)} total)")
print(f"Categorical features: {categorical_features}")


# Tạo pipeline cho xử lý cột số và cột hạng mục
# Cột số: Impute (điền NaN nếu còn sót) -> Scale (chuẩn hóa)
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')), # Điền median cho NaN còn sót
    ('scaler', StandardScaler())
])

# Cột hạng mục: Impute (điền giá trị phổ biến nhất) -> OneHotEncode
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False)) # sparse_output=False để dễ xem
])

# Kết hợp các pipeline bằng ColumnTransformer
preprocessor = ColumnTransformer([
    ('num', numerical_pipeline, numerical_features),
    ('cat', categorical_pipeline, categorical_features)
], remainder='passthrough') # Giữ lại các cột không được chỉ định (nếu có)

# Chia dữ liệu train/test
# Với dữ liệu chuỗi thời gian, không nên shuffle để tránh nhìn thấy tương lai
# Tuy nhiên, nếu mục tiêu không phụ thuộc mạnh vào thứ tự (ví dụ dự đoán cho một ngày bất kỳ), có thể shuffle
# Ở đây, ta giả sử dự đoán cho các giờ tiếp theo nên không shuffle.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False, stratify=None) # stratify=None vì shuffle=False

print(f"Kích thước X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"Kích thước y_train: {y_train.shape}, y_test: {y_test.shape}")

# Áp dụng preprocessor
# Fit trên train và transform cả train và test
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Lấy tên cột sau khi OneHotEncode (quan trọng để diễn giải model sau này)
try:
    feature_names_out = preprocessor.get_feature_names_out()
except AttributeError: # Cho scikit-learn phiên bản cũ hơn
    # Tạo tên cột thủ công (ít chính xác hơn nếu có 'remainder')
    feature_names_out = numerical_features
    if 'onehot' in preprocessor.named_transformers_['cat'].named_steps:
        ohe_categories = preprocessor.named_transformers_['cat'].named_steps['onehot'].categories_
        ohe_feature_names = []
        for i, cat_col in enumerate(categorical_features):
            for cat_val in ohe_categories[i]:
                ohe_feature_names.append(f"{cat_col}_{cat_val}")
        feature_names_out.extend(ohe_feature_names)
    else: # Nếu không có onehot (ít khả năng)
        feature_names_out.extend(categorical_features)


X_train_processed_df = pd.DataFrame(X_train_processed, columns=feature_names_out, index=X_train.index)
X_test_processed_df = pd.DataFrame(X_test_processed, columns=feature_names_out, index=X_test.index)

print("\nĐã hoàn thành Preprocessing.")
print(f"Kích thước X_train sau khi xử lý: {X_train_processed_df.shape}")
print("5 dòng đầu của X_train đã xử lý:")
print(X_train_processed_df.head())

print("\n--- Hoàn thành ---")
print("Dữ liệu đã sẵn sàng để đưa vào mô hình học máy.")
print("Biến X_train_processed_df, X_test_processed_df, y_train, y_test có thể được sử dụng.")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from lightgbm import LGBMClassifier # Cần cài đặt pip install lightgbm
# from xgboost import XGBClassifier # Hoặc XGBoost: pip install xgboost

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score # Để đánh giá chéo tốt hơn

# Khởi tạo các mô hình
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, solver='liblinear', random_state=42),
    "SVM (RBF Kernel)": SVC(probability=True, random_state=42), # probability=True để tính ROC AUC
    "KNN": KNeighborsClassifier(n_neighbors=5), # Có thể tinh chỉnh n_neighbors
    "LightGBM": LGBMClassifier(random_state=42)
    # "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42) # Nếu dùng XGBoost
}

results = {}

# Huấn luyện và đánh giá từng mô hình
print("\n--- Huấn luyện và Đánh giá các Mô hình ---")

for name, model in models.items():
    print(f"\nĐang huấn luyện mô hình: {name}...")

    # Huấn luyện mô hình
    model.fit(X_train_processed_df, y_train)

    # Dự đoán trên tập test
    y_pred = model.predict(X_test_processed_df)
    y_pred_proba = model.predict_proba(X_test_processed_df)[:, 1] # Lấy xác suất cho lớp dương (lớp 1)

    # Tính toán các chỉ số đánh giá
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Lưu kết quả
    results[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "ROC AUC": roc_auc
    }

    print(f"{name} - Hoàn thành.")
    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  F1 Score: {f1:.4f}")
    print(f"  ROC AUC: {roc_auc:.4f}")


# Hiển thị kết quả tổng hợp
print("\n--- Kết quả Đánh giá Tổng hợp ---")
results_df = pd.DataFrame(results).T # Chuyển vị để mô hình là index
print(results_df.sort_values(by="ROC AUC", ascending=False)) # Sắp xếp theo ROC AUC
```
